# 폴더 내 csv파일 합치기

```python
try:
    path = '원천데이터/'   #폴더 경로
    files = glob.glob(path + "*.csv") #파일 이름, 기본 인코딩 utf-8
    csv = pd.DataFrame()
    for file_name in files:
        df = pd.read_csv(file_name)
        csv = csv.append(df, ignore_index=True) #csv 파일 하나로 합치기
    print(csv)

except Exception as ex:
    print('error' + str(ex))
```
